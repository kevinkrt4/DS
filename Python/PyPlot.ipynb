# Pyplot Lines

## Number of people drowned by falling Into a Pool correlates with movies Nicholas Cage appeared in

www.tylervigen.com/spurious-correlations

```python
from matplotlib import pyplot as plt

# Initialize variables
years =  [1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007,2008, 2009]
movies =    [2, 2, 2, 3,1, 1, 2, 3, 4, 1, 4]
drownings = [109, 102, 102, 98, 85, 95, 96, 98, 123, 94, 102]

movies_scaling_factor = 10

# Apply scaling
scaled_movies = []
for i in movies:
    scaled_movies.append(i * movies_scaling_factor)
print (scaled_movies)


```

```python
# Show plot table
index = 0
print('Source Data'.center(30))
print('Years'.center(9), 'Movies'.center(9), 'Drownings'.center(9))
while index < len(years):    
    print (repr(years[index]).rjust(9), repr(movies[index]).rjust(9), repr(drownings[index]).rjust(9))
    index += 1
#print(repr(num).center(3))
#print (str(s).center(3))
#print (repr(s))
```

Table 1: Data values

### Plot on a single graph

```python
# Plot
plt.close('all')
plt.plot(years, scaled_movies)
plt.plot(years, drownings)
plt.title('Cage Movies Versus Drownings')
plt.xlabel('Year')
plt.xlabel('Year')
plt.legend(['movies', 'drownings']) 


# Show plots
plt.show()
```

Figure 1 Data Plot

# PyPlot Functions

## Linear
### f(x) = ax + b

```python
from matplotlib import pyplot as plt

my_domain = []
my_range=[]

a = 1  # 'a' is the slope of the line, sometimes referred to as 'm'
b = 0  # y intercept

for x in range(0, 10):
    my_domain.append(x)
print (my_domain)

for x in my_domain:
    my_range.append(a * x + b)
print (my_range)

plt.plot(my_domain, my_range)
   
plt.show()
```

## Exponential
### f(x) = a * e**x + b

```python
from matplotlib import pyplot as plt

my_domain = []
my_range=[]

a = 2
b = 20   # Appears to be the y intercept
e = 2    # 'e' must be greater than zero and not equal to 1

if e < 0 or e == 1:
    print ("Current value of e is %f. e must be greater than zero and != 1." % (e))

for x in range(0, 5):
    my_domain.append(x)
print (my_domain)

for x in my_domain:
    my_range.append(a * (e ** x) + b)
print (my_range)

plt.plot(my_domain, my_range)

plt.show()
```

## Quadratic
### f(x) = ax**2 + bx + c

```python
from matplotlib import pyplot as plt

my_domain = []
my_range=[]

a = 1
b = 0
c = 0

for x in range(-5, 6):
    my_domain.append(x)
print (my_domain)

for x in my_domain:
    my_range.append((a * (x ** 2)) + (b * x) + c)
    
print (my_range)

plt.plot(my_domain, my_range)

plt.show()
```

## Linear Absolute Value
### f(x) = a * | x + b | + c

```python
from matplotlib import pyplot as plt

my_domain = []
my_range=[]

a = 1
b = 0
c = 3

for x in range(-5, 6):
    my_domain.append(x)
print (my_domain)

for x in my_domain:
    my_range.append((a * abs(x + b)) + c)
    
print (my_range)

plt.plot(my_domain, my_range)

plt.show()
```
